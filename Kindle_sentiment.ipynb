```python
"""Best Practices 
 1.Preprocessing and Cleaning
 2.Train Test Splits
 3.BOW,TFIDF,Word2Vec
 4.Train ML algorithm """ 

#Load dataset
import pandas as pd
data=pd.read_csv('Kindle_dataset.csv')
data=data[['reviewText','rating']]
```

```python
##Checking missing values
data.isnull().sum()
```

```python
data['rating'].value_counts()
```

```python
##Preprocessing and Cleaning

##Positive revier->1  and negative review->0
data['rating']=data['rating'].apply(lambda x:0 if x<3 else 1)
```

```python
data['rating'].unique()
```

```python
data['rating'].value_counts()
```

```python
## 1.Lower all cases 
data['reviewText']=data['reviewText'].str.lower()
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')



##Removing url
data['reviewText']=data['reviewText'].apply(lambda x: re.sub(r'https?://\S+|www\.\S+|ftp://\S+', ' ', str(x)))

##Remove html tags
data['reviewText']=data['reviewText'].apply(lambda x: re.sub(r'<.*?>', ' ', str(x)))

##Removing Special Characters
data['reviewText']=data['reviewText'].apply(lambda x:re.sub(r'[^a-zA-Z0-9\s]', ' ', str(x)))

##Remove any additional space
data['reviewText']=data['reviewText'].apply(lambda x: re.sub(r'\s+', ' ', str(x)).strip())

##Removing stopwords
data['reviewText']=data['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y.lower() not in stopwords.words('english')]))
```

```python
data.head()
```

```python
## Lemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()]) 
```

```python
data['reviewText']=data['reviewText'].apply(lambda x:lemmatize_words(x))
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data['reviewText'],data['rating'],test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer()
X_train_bow=bow.fit_transform(X_train).toarray()
X_test_bow=bow.transform(X_test).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
X_train_tfidf=tfidf.fit_transform(X_train).toarray()
X_test_tfidf=tfidf.transform(X_test).toarray()
```

```python
X_train_bow
```

```python
## Naive Bayes Algorithm
from sklearn.naive_bayes import GaussianNB
nb_model_bow=GaussianNB().fit(X_train_bow,y_train)
nb_model_tfidf=GaussianNB().fit(X_train_tfidf,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
y_pred_bow=nb_model_bow.predict(X_test_bow)
```

```python
y_pred_tfidf=nb_model_tfidf.predict(X_test_tfidf)
```

```python
confusion_matrix(y_test,y_pred_bow)
```

```python
print("BOW Accuracy :",accuracy_score(y_test,y_pred_bow))
```

```python
confusion_matrix(y_test,y_pred_tfidf)
```

```python
print("Tfidf Accuracy :",accuracy_score(y_test,y_pred_tfidf))
```
